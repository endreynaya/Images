```python
def transformation(i,j,n):
    if(i%2==0 and j%2==0):
        a,b=i//2,j//2
    elif(i%2==0 and j%2==1):
        a,b=i//2,(n+j)//2
    elif(i%2==1 and j%2==0):
        a,b=(n+i)//2,j//2
    elif(i%2==1 and j%2==1):
        a,b=(n+i)//2,(n+j)//2
    return a,b

transformation(1,1,8)
```
